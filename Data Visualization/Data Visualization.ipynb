# Data Visualization

#### Data Visualization is a process of understanding the data in a more a graphical way with the help of plots, patterns, designs, correlations, etc.  There are multiple python libraries with the help of which data visualization can be performed such as matplotlib, seaborn, plotly, etc. This helps us to understand the data more clearly.

## Import Libraries 

```python
import pandas as pd
import numpy as np
import cv2
import os
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

## Content Images 

```python
Directory = r'D:\Style-Transfer\Webscrape\dataset\images\content_images'
Categories = []
```

```python
for category in os.listdir(Directory):
    Categories.append(category)
    
print(Categories)
```

```python
for category in Categories:
    folder = os.path.join(Directory,category)
    print(folder)
```

#### Displaying the total number of images in each category

```python
content_dict = {}

for category in Categories:
    folder = os.path.join(Directory,category)
    content_dict[category] = len(os.listdir(folder))
    print(category," ",len(os.listdir(folder)))
```

```python
from prettytable import PrettyTable

l = [str(content_dict['cars']), str(content_dict['museum']), str(content_dict['players'])]

table = PrettyTable(['cars', 'museum', 'players'])

table.add_row(l)
    
print(table)
```

```python
plt.bar(content_dict.keys(), content_dict.values(), color=(0.1, 0.1, 0.1, 0.1),  edgecolor='blue')
plt.title("Content Images")
plt.show()
```

#### Displaying one image from each category

```python
for category in Categories:
    folder = os.path.join(Directory,category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder,img)
        print(img_path)
        img_arr = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
        print("Image Shape: ",img_arr.shape)
        print("Image Size: ",img_arr.size)
        plt.imshow(img_arr[...,::-1])    # cv keeps the images in BGR
        plt.title(category)
        plt.show()
        break
```
